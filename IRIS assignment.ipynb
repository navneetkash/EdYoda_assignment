```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.feature_names)
```

```python
print(iris.target_names)
```

```python
print(iris.data[:5])
```

```python
x = iris.data
y = iris.target
```

```python
feature_data = pd.DataFrame(x)
feature_data.describe()
```

```python
# blue --> sepal lenght
# orange --> sepal width
# green --> petal length
# dark pink --> petal width
```

```python
plt.plot(feature_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.24,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
x_train_scaled = mm.fit_transform(X_train)
```

```python
x_test_scaled = mm.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = range(1,15)
total_acc = []
for i in k:
    flower = KNeighborsClassifier(n_neighbors=i)
    flower.fit(x_train_scaled,Y_train)
    acc = flower.score(x_test_scaled,Y_test)
    total_acc.append(acc)
```

```python
print(max(total_acc))
print(set(total_acc))
```

```python
plt.xlabel("value of K")
plt.ylabel("Accuracy")
plt.title("Select best value of k")
plt.plot(k,total_acc)
```
